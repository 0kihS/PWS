Eerst installeren we de benodigde bestanden, en downloaden we het model. Je kan dit uitvoeren met de run cell knop linksboven.

```python
!pip install unsloth
from unsloth import FastLanguageModel
from transformers import TextIteratorStreamer
from threading import Thread
import textwrap
max_print_width = 100

max_seq_length = 8192
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="lenarc/oreola", # Vervang dit door het model dat je wilt gebruiken
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit
)

text_streamer = TextIteratorStreamer(tokenizer)
```

Nu kunnen we tekst genereren! Vervang de tekst tussen aanhalingstekens hieronder met de tekst die je wilt laten afmaken. De lengte van de antwoorden kan je aanpassen door de max_new_tokens groter of kleiner te maken.

```python
inputs = tokenizer(
[
"""You are a helpful assistant
    Q: Hoe voel je je?
    A: Ik voel me zo: vogels zijn schattig!
    Q: Van wat voor vogels hou je?
    A:"""
]*1, return_tensors = "pt").to("cuda")

FastLanguageModel.for_inference(model)

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 80)
```
